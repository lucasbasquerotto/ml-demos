```python
%matplotlib inline

import gym
import numpy as np
from matplotlib import pyplot as plt
```

```python
env = gym.envs.make("MountainCar-v0", render_mode='rgb_array')
```

```python
env.reset()
plt.figure()
plt.imshow(env.render())

[env.step(0) for x in range(10000)]
plt.figure()
plt.imshow(env.render())

env.close()
```
